this project is to predict passengers who survived in Titanic and the common between them uing machine learning 

```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
pId= test_data["PassengerId"]
print(pId)
```

First of all i will clean the data by deciding which columns to drop based on the columns that will not affect my prediction

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data=train_data.drop(['Cabin','Name','Ticket'],axis='columns')
train_data.head()
```

Here i will replace the rows that have missing data with values 

```python
def replace(train_data):
    train_data.Fare.fillna(train_data.Fare.median(),inplace=True)
    train_data.Embarked.fillna("U",inplace=True)
    train_data.Sex.fillna("U",inplace=True)
    train_data.Age.fillna(train_data.Age.median(),inplace=True)
    train_data.Parch.fillna(train_data.Parch.median(),inplace=True)
    return train_data
    
```

```python
train_data=replace(train_data)
print(train_data)
```

After that i found aome columns containing strings so i will change them into numbers

```python
from sklearn.preprocessing import LabelEncoder

le_Sex=LabelEncoder()
le_Embarked=LabelEncoder()
train_data['Sex']= le_Sex.fit_transform(train_data['Sex'])
train_data['Embarked']=le_Embarked.fit_transform(train_data['Embarked'])
train_data.head()
```

```python
test_data=test_data.drop(['Cabin','Name','Ticket'],axis='columns')
test_data.head()
```

```python
def replace2(train_data):
    test_data.Fare.fillna(test_data.Fare.median(),inplace=True)
    test_data.Embarked.fillna("U",inplace=True)
    test_data.Sex.fillna("U",inplace=True)
    test_data.Age.fillna(test_data.Age.median(),inplace=True)
    test_data.Parch.fillna(test_data.Parch.median(),inplace=True)
    return test_data
```

```python
test_data=replace2(test_data)
print(test_data)
```

#convert letters into numbers to Sex and Embarked

```python
from sklearn.preprocessing import LabelEncoder

te_Sex=LabelEncoder()
te_Embarked=LabelEncoder()
test_data['Sex']= te_Sex.fit_transform(test_data['Sex'])
test_data['Embarked']=te_Embarked.fit_transform(test_data['Embarked'])
test_data.head()
```

```python
y=train_data["Survived"]
X=train_data.drop(['Survived','PassengerId'],axis='columns')

```

To validate the model i used test split

```python
from sklearn.model_selection import train_test_split
X, X_test ,y ,y_test =train_test_split(X, y , test_size=0.3,random_state=0)
```

i made the model  using RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)
clf.fit(X,y)
y_pred=clf.predict(X_test)
```

i checked the accuracy of the prediction

```python
from sklearn import metrics
print("Accuracy",metrics.accuracy_score(y_test,y_pred))
```
